```python
# 필요 패키지 전부 다
import os
import json
import requests

import beforehair1 as bh

# cpu만 강제 사용
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# 모듈 불러와서 무한루프 돌리기
while True:
    img_test = '홍길동_01012345678_man_cut.jpg'  # 이미지 이름이 이렇다는 가정 하에
    name = img_test.split('_',3)[0]
    name = name.split('.')[0]
    phone_number = '01012345678'

    # 사진 위치 경로 / 복제 시킬 경로
    base_path = './hair_image/before_hair/'
    storage_path = './hair_image/storage/'
    
    face_shape, hair_style, hair_length = bh.face_classify(name,phone_number,base_path,storage_path)
    
    filepath1 = './hair_image/recom_hair/male/Heart/cut/5/dandy/dandy1.jpg'
    filepath2 = './hair_image/recom_hair/male/Heart/cut/5/gail/gail1.jpg'
    filepath3 = './hair_image/recom_hair/male/Heart/cut/5/ivyleague/ivyleague1.jpg'
    filepath4 = './hair_image/recom_hair/male/Heart/cut/5/leaf/leaf1.jpg'
    
    # 사용자 정보 json 파일로 반환
    after_table = {'type':'bigdata',
                  'name':name,
                  'face_shape':face_shape,
                  'recom_hair1':filepath1,
                  'recom_hair2':filepath2,
                  'recom_hair3':filepath3,
                  'recom_hair4':filepath4,
                  'before_hair':hair_style,
                  'after_hair':'보브'}

    print(after_table)

    # 정보 전송
    headers = {'Content-Type': 'application/json; chearset=utf-8'}
    res = requests.post('http://was.t-seonghun.net:8000/insertdata/', data=json.dumps(after_table), headers=headers)
    
    with open('./hair_image/user_json/test.json', 'w', encoding='utf-8') as json_file:
          json.dump(after_table, json_file, ensure_ascii=False, indent='\t')

    break
```
