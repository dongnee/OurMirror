### 함수 생성

```python
# 전체 적용 함수

title_list = []
# date_list = []
post_link_list = []
img_link_list = []


def hair_crawling(base_url):
    # 헤더 및 파서 생성
    headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
    res = requests.get(base_url, headers=headers)
    html = res.text
    bs_obj = bs4.BeautifulSoup(html,'html.parser')
    
    # 가져올 것 : 글 제목, 날짜, 링크, 사진 링크
    title = bs_obj.findAll('a',{'class':'name_link _foryou_trigger'})
    post_num = len(title)
    
    for i in range(0, post_num):
        title_list.append(title[i].text)
        # date = bs_obj.findAll('span',{'class':'info'})[j].text
        link = bs_obj.findAll('a',{'class':'name_link _foryou_trigger'})[i]['href']
        post_link_list.append(link)
        image = bs_obj.findAll('div',{'class':'api_flicking_wrap thumb_group_flick_wrap _scroll_target'})[i]
        image_num = len(image)
        img_list = []
        for j in range(0, image_num):
            img_list.append(image.select('img')[j]['src'])
        img_link_list.append(img_list)
        
    hair_df = pd.DataFrame({'포스트 제목':title_list,'포스트 링크':post_link_list,'이미지 링크':img_link_list})
    return hair_df
```

```python
naver_hair = hair_crawling(base_url)
naver_hair
```

```python
naver_hair.to_csv('./data/네이버_유행스타일.csv')
```
