```python
import numpy as np
import matplotlib.pyplot as plt

# 그래프에서 한글 폰트가 깨질 경우 실행
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
fig = plt.figure(figsize=(20,6))
plt.box(False)

bg_color = '#000000'
fig.patch.set_facecolor(bg_color)
fig.set_facecolor(bg_color)

colors = ['#D9E5FF', '#B2CCFF', '#6799FF', '#4374D9', '#003399']
wedgeprops = {'width': 0.75, 'edgecolor': 'white', 'linewidth': 3}
textprops = {'fontsize': 12, 'color' : 'white'}

plt.pie(values, labels=names, autopct='%.1f%%',
        shadow=True, colors=colors, wedgeprops=wedgeprops, textprops=textprops)

# plt.legend(names) ## 범례 표시
# plt.legend(loc = 'upper right')
plt.show()
```
